```r
# ggplot definitions
colour_low <- "#db584a"
colour_mid <- "#ffffff"
colour_high <- "#36849c"
```

```r
data <- read.csv('/nfs/research1/zerbino/jhidalgo/inteql_GTEX_v8/data/processed_data/eqtl_gtex_corr.csv',header = TRUE)
```

```r
head(data)
```

```r
data <- melt(data,variable.name = 'Y',value.name = 'Pearson_correlation')
```

```r
data
```

```r
plot <- ggplot(
  data,
  aes(
      y=Y,
      x=X,
      fill=Pearson_correlation))+
#   geom_tile(
#     fill="grey95",
#     colour="white",
#     size=1)+
  geom_tile(
    aes(height=(sqrt(abs(Pearson_correlation/max(Pearson_correlation))))/1.1,width=(sqrt(abs(Pearson_correlation/max(Pearson_correlation))))/1.1))+
  scale_fill_gradient2(low = colour_low,mid=colour_mid,high = colour_high)+
  guides(
    fill=guide_legend(
      title="Pearson\ncorrelation"))+
#   labs(title="Relative \u0394Age, \u0394Height, \u0394Weight and \u0394Oxygen levels")+
  theme_heatmap()
```

```r
ggsave(filename = '/nfs/research1/zerbino/jhidalgo/inteql_GTEX_v8/data/processed_data/eqtl_gtex_corr.png',plot = plot,units = "cm",height = 30,width = 30)
```
