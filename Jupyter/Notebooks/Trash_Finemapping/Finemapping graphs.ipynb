```r
data=read.csv('eQTLs_after_finemapping.csv')
```

```r
data
```

```r
xmin=quantile(data$slope,0.01,na.rm = TRUE)
xmax=quantile(data$slope,0.99,na.rm = TRUE)
ggplot(data, aes(x=as.numeric(slope))) + 
    geom_histogram(aes(y=..density..),      # Histogram with density instead of count on y-axis
                   binwidth=.05,
                   colour="black", fill="white") +
    geom_density(alpha=.2, fill="#FF6666")+  # Overlay with transparent density plot
    theme_light()+
    xlim(xmin,xmax)
```
